##### Step 1: Create datasets in memory

```python
import random
from datetime import datetime, timedelta

# --- Dataset A: User Registrations 
names = ["alice", "bob", "charlie", "dave", "eve", "frank", "grace", "heidi"]
dataset_a = []

for i in range(8):
    username = names[i]
    email = f"{username}@example.com"
    age = random.randint(18, 60)
    signup_date = (datetime(2025,1,1) + timedelta(days=random.randint(0,30))).strftime("%Y-%m-%d")
    
    dataset_a.append({
        "username": username,
        "email": email,
        "age": age,
        "signup_date": signup_date
    })

# Adding some invalid entries
dataset_a[1]["email"] = "bobexample.com"       
dataset_a[2]["username"] = ""                 
dataset_a[3]["age"] = "twenty"                 

print("Dataset A:", dataset_a)
```

```python
# --- Dataset B: Application Logs 
services = ["auth", "payment", "search"]
levels = ["INFO", "WARNING", "ERROR"]
messages = [
    "Operation successful", "Operation failed", "Timeout occurred",
    "User login", "Data saved", "Cache cleared", "Invalid request"
]

dataset_b = []
base_time = datetime(2025,1,10,8,0,0)
for i in range(10):
    timestamp = (base_time + timedelta(minutes=i*5)).strftime("%Y-%m-%d %H:%M:%S")
    service = random.choice(services)
    level = random.choice(levels)
    message = random.choice(messages)
    
    dataset_b.append({
        "timestamp": timestamp,
        "service": service,
        "level": level,
        "message": message
    })

print("\nDataset B:", dataset_b)
```

##### Step 3: 
Write a validation function called validate_registrations(entries) that takes Dataset A and returns two lists: valid_entries and invalid_entries. A valid entry must have a non-empty username, an email containing @, and an age that can be converted to a positive integer. Test your function on Dataset A and print the counts.

```python
def validate_registrations(entries):
    valid_entries = []
    invalid_entries = []

    for entry in entries:
        username=entry.get("username","").strip()
        email=entry.get("email", "")
        age=entry.get("age", None)

        #checking
        is_valid_username=bool(username)
        is_valid_email="@" in email

        try:
            age_int=int(age)
            is_valid_age= age>0
        
        except:

            is_valid_age=False

        if is_valid_username and is_valid_email and is_valid_age:
            valid_entries.append(entry)

        else:
            invalid_entries.append(entry)

    return valid_entries, invalid_entries
```
